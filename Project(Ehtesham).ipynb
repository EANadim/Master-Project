# Check the Normal Distribution of Each Case (Human Match)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def remove_outliers_percentile(df, column, lower_percentile=5, upper_percentile=95):
    """ Remove outliers based on lower and upper percentile thresholds. """
    lower_bound = df[column].quantile(lower_percentile / 100)
    upper_bound = df[column].quantile(upper_percentile / 100)
    
    # Identify outliers
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    
    # Filter out outliers
    df_filtered = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    
    before_count = len(df)
    after_count = len(df_filtered)
    outliers_removed = before_count - after_count
    
    return df_filtered, outliers_removed, outliers[['participant_id', column]]

def plot_distribution(df, column):
    """ Plot histogram with normal distribution line and Q-Q plot. """
    plt.figure(figsize=(12, 6))
    
    # Histogram with normal distribution overlay
    plt.subplot(1, 2, 1)
    n, bins, patches = plt.hist(df[column], bins=30, edgecolor='k', alpha=0.7, density=True)
    mu, std = np.mean(df[column]), np.std(df[column])
    best_fit_line = stats.norm.pdf(bins, mu, std)
    plt.plot(bins, best_fit_line, 'r--', linewidth=2)
    plt.title('Histogram with Normal Distribution')
    plt.xlabel(column)
    plt.ylabel('Density')
    
    # Q-Q Plot
    plt.subplot(1, 2, 2)
    stats.probplot(df[column], dist="norm", plot=plt)
    plt.title('Q-Q Plot')
    
    plt.tight_layout()
    plt.show()

def test_normality(df, column):
    """ Perform Shapiro-Wilk Test for normality. """
    stat, p_value = stats.shapiro(df[column])
    return stat, p_value

def print_shapiro_test(df, label):
    """ Print results of Shapiro-Wilk Test. """
    stat, p_value = test_normality(df, 'social_behavior_percentage')
    print(f"Shapiro-Wilk Test for {label}: Statistic={stat}, p-value={p_value}")

# Example DataFrame (replace with your actual DataFrame)
# df = pd.DataFrame({
#     'fitting_status': ['Match', 'Match', 'Mismatch', 'Mismatch'],
#     'target_of_interaction': ['Human', 'Robot', 'Human', 'Robot'],
#     'social_behavior_percentage': [10, 95, 100, 105],
#     'participant_id': [1, 2, 3, 4]
# })

# Filter the rows
filtered_hm = df[(df['fitting_status'] == 'Match') & (df['target_of_interaction'] == 'Human')]
filtered_rm = df[(df['fitting_status'] == 'Match') & (df['target_of_interaction'] == 'Robot')]
filtered_hmis = df[(df['fitting_status'] == 'Mismatch') & (df['target_of_interaction'] == 'Human')]
filtered_rmis = df[(df['fitting_status'] == 'Mismatch') & (df['target_of_interaction'] == 'Robot')]

# Remove outliers based on percentiles and get the number of outliers removed
filtered_hm, hm_outliers_removed, hm_outliers = remove_outliers_percentile(filtered_hm, 'social_behavior_percentage')
filtered_rm, rm_outliers_removed, rm_outliers = remove_outliers_percentile(filtered_rm, 'social_behavior_percentage')
filtered_hmis, hmis_outliers_removed, hmis_outliers = remove_outliers_percentile(filtered_hmis, 'social_behavior_percentage')
filtered_rmis, rmis_outliers_removed, rmis_outliers = remove_outliers_percentile(filtered_rmis, 'social_behavior_percentage')

# Print the number of outliers removed and their participant_Id
print(f"Number of outliers removed from filtered_hm: {hm_outliers_removed}")
print(f"Participant Ids of outliers in filtered_hm:\n{hm_outliers}")

print(f"Number of outliers removed from filtered_rm: {rm_outliers_removed}")
print(f"Participant Ids of outliers in filtered_rm:\n{rm_outliers}")

print(f"Number of outliers removed from filtered_hmis: {hmis_outliers_removed}")
print(f"Participant Ids of outliers in filtered_hmis:\n{hmis_outliers}")

print(f"Number of outliers removed from filtered_rmis: {rmis_outliers_removed}")
print(f"Participant Ids of outliers in filtered_rmis:\n{rmis_outliers}")

# Plot distributions
plot_distribution(filtered_hm, 'social_behavior_percentage')
plot_distribution(filtered_rm, 'social_behavior_percentage')
plot_distribution(filtered_hmis, 'social_behavior_percentage')
plot_distribution(filtered_rmis, 'social_behavior_percentage')

# Perform Shapiro-Wilk Test
print_shapiro_test(filtered_hm, 'filtered_hm')
print_shapiro_test(filtered_rm, 'filtered_rm')
print_shapiro_test(filtered_hmis, 'filtered_hmis')
print_shapiro_test(filtered_rmis, 'filtered_rmis')
```

# Check the Normal Distribution of Each Case (Robot Match)

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import pandas as pd

# Example data for demonstration purposes
# Replace this with loading your actual DataFrame
# filtered_hmis = pd.read_csv('path_to_your_file.csv')

# Select the 'participant_id' and 'social_behavior_percentage' columns for analysis
data_to_test = filtered_rm[['participant_id', 'social_behavior_percentage']]

# Calculate the z-scores of 'social_behavior_percentage'
data_to_test['z_score'] = stats.zscore(data_to_test['social_behavior_percentage'])

# Identify outliers based on z-score (values outside of 2 standard deviations)
outliers = data_to_test[(data_to_test['z_score'] < -2) | (data_to_test['z_score'] > 2)]

# Print the IDs of outliers
print("Participant IDs of outliers:")
print(outliers['participant_id'].tolist())

# Optionally, print the number of outliers
print(f"\nNumber of outliers: {outliers.shape[0]}")

# Drop the z_score column for further analysis
data_to_test = data_to_test.drop(columns=['z_score'])

# Visualize the distribution of 'social_behavior_percentage'
plt.figure(figsize=(14, 6))

# Histogram and KDE plot
plt.subplot(1, 2, 1)
sns.histplot(data_to_test['social_behavior_percentage'], kde=True)
plt.title('Histogram and KDE Plot')
plt.xlabel('Social Behavior Percentage')

# Q-Q plot
plt.subplot(1, 2, 2)
stats.probplot(data_to_test['social_behavior_percentage'], dist="norm", plot=plt)
plt.title('Q-Q Plot')

plt.tight_layout()
plt.show()

# Perform normality tests
print("\nNormality Test Results:")

# Shapiro-Wilk test
shapiro_test = stats.shapiro(data_to_test['social_behavior_percentage'])
print(f"Shapiro-Wilk Test: Statistic={shapiro_test.statistic:.4f}, p-value={shapiro_test.pvalue:.4f}")

# Kolmogorov-Smirnov test (with a normal distribution)
ks_test = stats.kstest(data_to_test['social_behavior_percentage'], 'norm', args=(data_to_test['social_behavior_percentage'].mean(), data_to_test['social_behavior_percentage'].std()))
print(f"Kolmogorov-Smirnov Test: Statistic={ks_test.statistic:.4f}, p-value={ks_test.pvalue:.4f}")

# Interpretation of normality test results
alpha = 0.05
if shapiro_test.pvalue < alpha:
    print("Shapiro-Wilk Test: The data is not normally distributed (reject H0).")
else:
    print("Shapiro-Wilk Test: The data is normally distributed (fail to reject H0).")

if ks_test.pvalue < alpha:
    print("Kolmogorov-Smirnov Test: The data is not normally distributed (reject H0).")
else:
    print("Kolmogorov-Smirnov Test: The data is normally distributed (fail to reject H0).")
```

# Spearman Correlation (Proximity Distance)

```python
import pandas as pd
import scipy.stats as stats

# Load the Excel file
file_path = 'behavior_report_ehtesham.xlsx'  # Replace with your file path
df = pd.read_excel(file_path)

# Filter the rows where "fitting_status" is "Match" and "target_of_interaction" is "Human" or "Robot"
filtered_hm = df[(df['fitting_status'] == 'Match') & (df['target_of_interaction'] == 'Human')]
filtered_rm = df[(df['fitting_status'] == 'Match') & (df['target_of_interaction'] == 'Robot')]
filtered_hmis = df[(df['fitting_status'] == 'Mismatch') & (df['target_of_interaction'] == 'Human')]
filtered_rmis = df[(df['fitting_status'] == 'Mismatch') & (df['target_of_interaction'] == 'Robot')]

# Define the ordinal encoding
ordinal_encoding = {'Close': 3, 'Mid': 2, 'Far': 1}

# Encode ordinal data
encoded_ordinal_datahm = [ordinal_encoding[item] for item in filtered_hm['sitting_distance']]
encoded_ordinal_datarm = [ordinal_encoding[item] for item in filtered_rm['sitting_distance']]
encoded_ordinal_datahmis = [ordinal_encoding[item] for item in filtered_hmis['sitting_distance']]
encoded_ordinal_datarmis = [ordinal_encoding[item] for item in filtered_rmis['sitting_distance']]

# Continuous data for social interaction
social_interaction_hm = filtered_hm['social_behavior_percentage']
social_interaction_hmis = filtered_hmis['social_behavior_percentage']
social_interaction_rm = filtered_rm['social_behavior_percentage']
social_interaction_rmis = filtered_rmis['social_behavior_percentage']

# Calculate Spearman's rank correlation
spearman_corrhm, spearman_p_valuehm = stats.spearmanr(encoded_ordinal_datahm, social_interaction_hm)
spearman_corrrm, spearman_p_valuerm = stats.spearmanr(encoded_ordinal_datarm, social_interaction_rm)
spearman_corrhmis, spearman_p_valuehmis = stats.spearmanr(encoded_ordinal_datahmis, social_interaction_hmis)
spearman_corrrmis, spearman_p_valuermis = stats.spearmanr(encoded_ordinal_datarmis, social_interaction_rmis)

# Print the results
print(f"Spearman's rank correlation for Human Match: {spearman_corrhm}, p-value: {spearman_p_valuehm}")
print(f"Spearman's rank correlation for Robot Match: {spearman_corrrm}, p-value: {spearman_p_valuerm}")
print(f"Spearman's rank correlation for Human Mismatch: {spearman_corrhmis}, p-value: {spearman_p_valuehmis}")
print(f"Spearman's rank correlation for Robot Mismatch: {spearman_corrrmis}, p-value: {spearman_p_valuermis}")
```
